```r

library('readr')
library('plyr')
library('ggfortify')
setwd("/home/nazarov_group/CMV_project/AK_folder/cmv_vgenes/")
folder = "/home/nazarov_group/CMV_project/AK_folder/cmv_vgenes/"
folder1 = "/home/nazarov_group/CMV_project/AK_folder/cmv_jgenes/"
file_list = list.files(path = folder, pattern="HIP*")
remove_char = function(df) {
    #reading csv-files from dir -> removing 'unresolved' -> removing last 3 letters (isoforms) -> writing to new dir
    output_path = paste(folder1, df, sep = "")
    df_df = get(df)
    df_df = subset(df_df, J != 'unresolved')
    j_new = gsub('.{0,3}$', '', df_df$J)
    df_df$J = j_new
    write.csv(df_df, file = output_path)
}
for (i in 1:length(file_list)){
    assign(file_list[i],  
  read_csv(file_list[i], col_names = TRUE))
}
df_fin = lapply(file_list, remove_char)
```

```r
setwd("/home/nazarov_group/CMV_project/AK_folder/cmv_jgenes")
folder = "/home/nazarov_group/CMV_project/AK_folder/cmv_jgenes"
file_list = list.files(path = folder, pattern="HIP*")
folder1 = '/home/nazarov_group/CMV_project/AK_folder/cmv_jgenes_freq/'
vgene_fun = function(df) {
    # reading csv-files from dir -> counting frequences -> transposing table -> writing to new dir
    df_df = get(df)
    j_gene_count = count(df_df, 'J')
    t_j_gene_count = t(j_gene_count)
    newcol = c(df)
    t_j_gene_count <- cbind(newcol, t_j_gene_count)
    }
for (i in 1:length(file_list)){
    assign(file_list[i],  
  read_csv(file_list[i]))
}
entropy_table = sapply(file_list, vgene_fun)
for (i in 1:length(entropy_table)) {
df1 = as.data.frame(entropy_table[i])
colnames(df1) <- as.character(unlist(df1[1,]))
df1 = df1[-1, ]
df1_name = colnames( df1 )[1]
row.names(df1) <-  colnames( df1 )[1]
df1 <- df1[,-1]
output_path = paste(folder1, df1_name, sep = "")
write.csv(df1, file = output_path)
    }
```
