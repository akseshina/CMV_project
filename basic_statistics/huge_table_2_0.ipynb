```r
library('readr')
library('plyr')
setwd("/home/nazarov_group/CMV_project/AK_folder")
folder = "/home/nazarov_group/CMV_project/AK_folder"
file_list = list.files(path = folder, pattern="HIP*")
vgene_fun = function(df) {
    df_df = get(df)
    v_gene_count = count(df_df, 'V')
    t_v_gene_count = t(v_gene_count)
    newcol = c(df)
    t_v_gene_count <- cbind(newcol, t_v_gene_count)
    }
for (i in 1:length(file_list)){
    assign(file_list[i],  
  read_csv(file_list[i]))
}
entropy_table = sapply(file_list, vgene_fun)
df1 = as.data.frame(entropy_table[1])
colnames(df1) <- as.character(unlist(df1[1,]))
df1 = df1[-1, ]
row.names(df1) <-  colnames( df1 )[1]
df1 <- df1[,-1]
df1
df2 = as.data.frame(entropy_table[2])
colnames(df2) <- as.character(unlist(df2[1,]))
df2 = df2[-1, ]
row.names(df2) <-  colnames( df2 )[1]
df2 <- df2[,-1]
df1[setdiff(names(df2), names(df1))] <- NA
df2[setdiff(names(df1), names(df2))] <- NA
df = rbind(df1, df2)
df
```
