```python
max_temp_mass=[]
sum_associated_temp=[]
for index,row in metadata.iterrows():
    #making string for human
    ##bug string = metadata.iloc[i]["sample_id"]
    ##
    stro=row["sample_id"]
    dstring='~/data/cmv/'+stro+'.tsv' #organization of reading
    print(dstring)
    cmv_temporary=pd.read_csv(dstring,sep=',')
    
    #formation of true/false massiv
    temp_temporary=cmv_temporary['amino_acid'].isin(cmv_associated_de_novo_164['CDR3'])
    
    #n of cmv-associated de novo tcr on human
    temp_sum=cmv_temporary['amino_acid'].isin(cmv_associated_de_novo_164['CDR3']).sum()
    print(temp_sum)
    #шпора temp_sort_positive_HIP08653= cmv_positive_HIP08653[true_false_cmv_positive_HIP08653]
    #cоздание датасета из загруженного .где сmv, определенное только что - true
    templates_cmv=cmv_temporary[temp_temporary]

    #max templates of cmv-associated tcr in human, this is row with info
    ##temp_max=temp_sort_negative.loc[temp_sort_negative['templates'].idxmax()]
    #row_temp_max=templates_cmv.loc[templates_cmv['templates'].idxmax()]
    #print(row_temp_max)
    #df['Age'].max()
    #max - int, I will put it to mass later
    temp_max=templates_cmv['templates'].max()
    print(temp_max)
    max_temp_mass.append(temp_max)
    sum_associated_temp.append(temp_sum)
    
print(max_temp_mass) 
```
